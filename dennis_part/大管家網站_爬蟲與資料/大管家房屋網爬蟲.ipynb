```python
with open ('data' + '.txt','w') as f:
    for page in range(1,99):
        r = requests.get('http://www.chrb.com.tw/tenement.php?&pages=' + str(page))
        r.encoding='utf-8'
        content = r.text
        soup = BeautifulSoup(content,'lxml')
        info = soup.findAll('td',class_="tenement_07")
        whos = soup.findAll('span' , class_ = "tenement_04")
        i = 0
        record = 0
        sum = ''
        for u in info:
            record +=1
            cur = u.get_text()
            sum += cur
            i += 1
            if i>=3 and record < 84:
                want = sum + whos[int(record/3)].get_text() + '\n'
                f.write(want)
                sum =""
                i =0
```
