```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
link = 'https://rent.591.com.tw/?kind=0&region=4'
headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_12_3) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/56.0.2924.87 Safari/537.36'}
data = {'MIME 類型': 'application/x-www-form-urlencoded',
'title': "【台北市出租】-591房屋交易網",
"device": "pc"}
r = requests.get(link ,headers = headers)
content = r.text
soup = BeautifulSoup(content,'lxml')
```

```python
info = soup.findAll(class_ = "pull-left infoContent")
ls = []
titles = []
for im in info:
    pattern = im.p.get_text()
    p = im.a.get_text()
    titles.append(p)
#     pattern.replace(' ' , '')
    pattern = "".join(pattern.split())
    ls.append(pattern)
ls = pd.DataFrame(ls,columns=["規格"])
titles = pd.DataFrame(titles,columns=['名稱'])
pri_ls = []
divs = soup.findAll(class_ = 'price')
for div in divs:
    price = div.i.get_text()
    pri_ls.append(price)
pri_df = pd.DataFrame(pri_ls,columns=['價錢'])
data = pd.concat([ls,titles,pri_df],axis = 1)
data.head()
```
